```python
import numpy as np
import pandas as pd
```

```python
#reading the excel sheet1
purchase_data = pd.read_excel(r"C:\Users\Admin\Downloads\Lab Session1 Data.xlsx")
purchase_data
```

```python
#getting the dimension of the dataset
purchase_data.shape
```

```python
#making the dataset into a datframe and thus removing the NaN values using dropna
purchase_dataframe = pd.DataFrame(purchase_data)
purchase_dataframe1 =purchase_dataframe.dropna(axis = 1)
print(purchase_dataframe1)
```

```python
#getting the attributes of AX=C
matrix_A = purchase_dataframe1.iloc[:: , 1:4]
print(matrix_A)
matrix_A.shape
```

```python
#getting the output of the matrix AX=C
matrix_C = purchase_dataframe1.iloc[:: , 4:5]
print(matrix_C)
matrix_C.shape
```

```python
#finding rank of the attribute matrix
rank_A = np.linalg.matrix_rank(matrix_A)
print(rank_A)
```

```python
#finding rank of the output matrix
rank_C = np.linalg.matrix_rank(matrix_C)
print(rank_C)
```

```python
#pseudo inverse of attribute matrix as it isn't a square matrix
pseudo_inv_A = np.linalg.pinv(matrix_A)
print(pseudo_inv_A)
```

```python
#pseudo inverse of output matrix as it isn't a square matrix
pseudo_inv_C = np.linalg.pinv(matrix_C)
print(pseudo_inv_C)
```

```python
#multiplying the pseudo inverse of attribute matrix with output matrix in order to get X of AX=C
matrix_X = np.dot(pseudo_inv_A, matrix_C)
print(matrix_X)
```

```python
#adding an extra column for classifying the customers as Poor or Rich acc to their payment
matrix_C.loc[matrix_C['Payment (Rs)'] <= 200, 'Classification'] = 'Poor' 
matrix_C.loc[matrix_C['Payment (Rs)'] > 200, 'Classification'] = 'Rich' 
print(matrix_C)
```

```python
#reading the 2nd sheet of the excel doc
stock_price = pd.read_excel(r"C:\Users\Admin\Downloads\Lab Session1 Data.xlsx" , sheet_name = 'IRCTC Stock Price')
stock_price
```

```python
#finding mean/avg acc to the price column 
price_mean = stock_price[["Price"]].mean()
print("The population mean is: ",price_mean)
```

```python
#finding variance acc to the price column
price_variance = stock_price[["Price"]].var()
print("The population variance is: ",price_variance)
```

```python
#filtering out all Wednesdays
stock_wed = stock_price.loc[stock_price['Day'] == 'Wed']
stock_wed
```

```python
#finding mean of all the Wednesdays from the filtered df
stock_wed_mean  = stock_wed[["Price"]].mean()
print("The sample mean for all Wednesdays: ", stock_wed_mean)
```

```python
#finding variance of all the Wednesdays from the filtered df
stock_wed[["Price"]].var()
```

```python
#comparing the sample mean of all the wednesdays with the population mean
print("The difference between the sample mean and the population mean is: ")
print(abs(stock_wed_mean - price_mean))
```
