```python
import numpy as np
import pandas as pd
```

```python
#loading the dataset
documents = pd.read_excel(r"C:\Users\Admin\Downloads\Lab Session2 Data.xlsx")
documents
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
#tokenization
tokens = documents.apply(lambda row: nltk.word_tokenize(row['Report']), axis=1)
tokens
```

```python
#token_population
token_population = []
for i in range(len(tokens)):
    for j in tokens[i]:
        token_population.append(j)
print(token_population)
print(len(token_population))
```

```python
#converting the token population into all lower cases so we can eliminate more no.of stopwords
token_population_low = []
for i in token_population:
    token_population_low.append(i.lower())
print(token_population_low)
print(len(token_population_low))
    
```

```python
#removing stopwords and getting a more accurate result after making all words lowercase
nltk.download('stopwords')
from nltk.corpus import stopwords
bag_of_words = []
stop_words = set(stopwords.words('english'))
for i in token_population_low:
    if i not in stop_words:
        bag_of_words.append(i)
print(bag_of_words)
print(len(bag_of_words))
```

```python
#removing all the duplicates in the bag of words
bagofwords = []
bagofwords = list(set(bag_of_words))
print(bagofwords)
print(len(bagofwords))
```

```python
#creating the first vector
v1 = []
for i in tokens:
    v1_temp = []
    for j in bagofwords:
        if(j in i):
            v1_temp.append(1)
        else:
            v1_temp.append(0)
    v1.append(v1_temp)
documents['v1_bagofwords'] = v1  
documents
```

```python
#creating the second vector
v2 = []
for i in tokens:
    v2_temp = []
    for j in bagofwords:
        v2_temp.append(i.count(j))
    v2.append(v2_temp)
for i in v2:
    print(i)
```

```python
#for jaccard coeff as well as the jaccard distance

#importing the required packages
from scipy.spatial.distance import jaccard
from sklearn.metrics import jaccard_score

#calculating the jaccard similarity coeff as well as the distance for a particular document
vectors = documents['v1_bagofwords']
similarity = jaccard_score(v1[0], v1[3])
distance = jaccard(v1[0], v1[3])
print("Jaccard similarity is equal to: ", similarity)
print("Jaccard distance is equal to: ", distance)
```

```python
jaccard_matrix=[[0 for columns in range(37)] for rows in range(37)]

for i in range(37):
    for j in range(37):
        similarity = jaccard_score(v1[i],v1[j])
        jaccard_matrix[i][j] = similarity
print(jaccard_matrix)
```

```python
#importing the required library
from numpy.linalg import norm

#assigning 2 particular v2 vectors 
A = v2[0]
B = v2[3]
print("A:", A)
print("B:", B)

#computing cosine similarity
cosine = np.dot(A, B)/(norm(A)*norm(B))
print("Cosine Similarity:", cosine)
```

```python
cosine_matrix=[[0 for columns in range(37)] for rows in range(37)]

for i in range(37):
    for j in range(37):
        cosine = np.dot(v2[i],v2[j])/(norm(v2[i])*norm(v2[j]))
        cosine_matrix[i][j] = cosine
print(cosine_matrix)
```
